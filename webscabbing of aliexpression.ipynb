```python

product_names = []
prices = []
star_ratings = []
previous_prices = []
store_links = []
discounts = []
shipping_statuses = []
totalsolds = []
ratings = []

totalpage = 10
for page in range(1, totalpage + 1):
    url = f"https://www.aliexpress.com/w/wholesale-phone.html?page={page}&g=y&SearchText=phone"
    driver.get(url)
    soup = BeautifulSoup(driver.page_source, 'html.parser')
    totalphone = soup.find_all("div", class_="list--gallery--C2f2tvm search-item-card-wrapper-gallery")

    for phone in totalphone:
        product_name = phone.find("div", class_="multi--title--G7dOCj3").text.strip() if phone.find("div", class_="multi--title--G7dOCj3") else ''
        price = phone.find("div", class_="multi--price-sale--U-S0jtj").text.strip().replace('HUF', '').replace(',', '').replace('.', '') if phone.find("div", class_="multi--price-sale--U-S0jtj") else ''
        totalsold = phone.find('span', class_="multi--trade--Ktbl2jB").text.strip() if phone.find('span', class_="multi--trade--Ktbl2jB") else ''
        previous_price = phone.find('div', class_='multi--price-original--1zEQqOK').text.strip().replace('HUF', '').replace(',', '').replace('.', '') if phone.find('div', class_='multi--price-original--1zEQqOK') else ''
        shipping_status = phone.find("div", class_="multi--serviceContainer--3vRdzWN").text.strip() if phone.find("div", class_="multi--serviceContainer--3vRdzWN") else 'No shipping'
        
        try:
            rating = phone.find('div', class_='multi--starList--t9_CAj2').text.strip()
        except AttributeError:
            rating = ''
        
        try:
            store_link = phone.find("a", class_="cards--storeLink--XkKUQFS")['href']
        except (TypeError, KeyError):
            store_link = ''
        
        try:
            store_name = phone.find("a", class_="cards--storeLink--XkKUQFS").text.strip()
        except AttributeError:
            store_name = ''
        
        try:
            discount = phone.find("span", class_="multi--discount--3hksz5G").text.strip()
        except AttributeError:
            discount = ''
        try:
            star_rating = phone.find("div", class_="multi--starList--t9_CAj2").text.strip()
        except AttributeError:
            star_rating = ''
        
        # Append data to respective lists
        product_names.append(product_name)
        prices.append(price)
        star_ratings.append(store_name)
        previous_prices.append(previous_price)
        store_links.append(store_link)
        discounts.append(discount)
        shipping_statuses.append(shipping_status)
        totalsolds.append(totalsold)
        ratings.append(rating)

# Quit the browser
driver.quit()

# Create DataFrame
data = {
    'product_name': product_names,
    'price': prices,
    'star_rating': item_ratings,
    'previous_price': previous_prices,
    'store_link': store_links,
    'discount': discounts,
    'shipping_status': shipping_statuses,
    'totalsold': totalsolds,
    'rating': ratings
}

df = pd.DataFrame(data)
print(df.head())
```
